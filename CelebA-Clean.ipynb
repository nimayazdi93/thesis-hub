```python
import pandas as pd
import numpy as np
import shutil
MasterDataFrame=pd.read_csv('data-set/celeba/identity_CelebA.csv')
MasterDataFrame
```

```python
Indentities=pd.DataFrame(MasterDataFrame['Identity'].drop_duplicates())
Indentities
```

```python
Indentities=Indentities.iloc[np.random.permutation(len(Indentities))].head(200)
Indentities
```

```python
NewMasterDataFrame=pd.DataFrame(columns=['Image','Identity'])
for id in Indentities['Identity']:
    ImageForThisID=MasterDataFrame[MasterDataFrame['Identity']==id].head(10)
    NewMasterDataFrame=NewMasterDataFrame.append(ImageForThisID)
```

```python
NewMasterDataFrame.to_csv('data-set/celeba/Clean/Identity_CelebA.csv')
```

```python
for Image in NewMasterDataFrame['Image']:
    shutil.copy('data-set/celeba/images/'+Image, 'data-set/celeba/Clean/'+Image)
```
